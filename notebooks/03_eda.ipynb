```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re

from sklearn.preprocessing import MinMaxScaler

from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('../dataset/preprocessed_train_data.csv')
data = data.drop(columns='name')
```

```python
X = data.drop(columns='transported')
y = data['transported']
```

```python
scaler = MinMaxScaler()
scaler.fit_transform(X)
```

```python
data_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
data_scaled
```

```python
model = RandomForestClassifier()
model.fit(data_scaled, y)
```

```python
model.score(data_scaled, y)
```

```python
test_data = pd.read_csv('../dataset/test.csv')
```

```python
test_data
```

```python
test_data.columns = [re.sub(r'(?<!^)(?=[A-Z])', '_', col).lower() for col in test_data.columns]
```

```python
test_data = test_data.drop(columns='passenger_id')
```

```python
cryo_sleep_encoded = pd.get_dummies(test_data['cryo_sleep'].fillna('Unknown'), prefix='cryo_sleep').astype(int)
test_data = pd.concat([test_data.drop(columns='cryo_sleep'), cryo_sleep_encoded], axis=1)

test_data[['deck', 'num', 'side']] = test_data['cabin'].str.split('/', expand=True)
test_data = test_data.drop(columns='cabin')

data_deck_encoded = pd.get_dummies(test_data['deck'].fillna('Unknown'), prefix='deck').astype(int)
test_data = pd.concat([test_data.drop(columns='deck'), data_deck_encoded], axis=1)

data_side_encoded = pd.get_dummies(test_data['side'].fillna('Unknown'), prefix='side').astype(int)
test_data = pd.concat([test_data.drop(columns='side'), data_side_encoded], axis=1)

test_data['destination'] = test_data['destination'].fillna('Unknown')
data_destination_encoded = pd.get_dummies(test_data['destination'], prefix='destination').astype(int)
test_data = pd.concat([test_data.drop(columns='destination'), data_destination_encoded], axis=1)
```
