# Spaceship titanic - Kaggle competition

#### Competition link: https://www.kaggle.com/competitions/spaceship-titanic/overview

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../dataset/train.csv')
data
```

```python
data.columns = data.columns.str.replace(r'(?<!^)(?=[A-Z])', '_', regex=True).str.lower()
data
```

```python
data['home_planet'] = data['home_planet'].fillna('unknown_planet')
data_home_planet = pd.get_dummies(data['home_planet'], drop_first=True).astype(int)
data = pd.concat([data.drop(columns='home_planet'), data_home_planet], axis=1)
```

```python
data['cryo_sleep'] = data['cryo_sleep'].fillna('unknown')
data_cryo_sleep = pd.get_dummies(data['cryo_sleep'], drop_first=True, prefix='cryo_sleep').astype(int)
data = pd.concat([data.drop(columns='cryo_sleep'), data_cryo_sleep], axis=1)
data
```

```python
data[['deck', 'num', 'side']] = data['cabin'].str.split('/', expand=True)
data = data.drop(columns='cabin')
```

```python
data
```

```python
data['destination'] = data['destination'].fillna('unknown')
data_home_planet = pd.get_dummies(data['destination'], drop_first=True, prefix='planet').astype(int)
data = pd.concat([data.drop(columns='destination'), data_home_planet], axis=1)
```

```python
data = data.rename(columns={
    'PSO J318.5-22': 'pso_j3128_5_22',
    'TRAPPIST-1e': 'trappist_1e'
})
```

```python
data
```

```python
data['deck'] = data['deck'].fillna('unknown')
data_deck = pd.get_dummies(data['deck'], drop_first=True, prefix='deck').astype(int)
data = pd.concat([data.drop(columns='deck'), data_deck], axis=1)

data['side'] = data['side'].fillna('unknown')
data_deck = pd.get_dummies(data['side'], drop_first=True, prefix='side').astype(int)
data = pd.concat([data.drop(columns='side'), data_deck], axis=1)
```

```python
data['age'] = pd.to_numeric(data['age'], errors='coerce')
data['age'] = data['age'].fillna(data['age'].median())
```

```python
data['v_i_p'] = data['v_i_p'].fillna('unknown')
data_deck = pd.get_dummies(data['v_i_p'], drop_first=True, prefix='v_i_p').astype(int)
data = pd.concat([data.drop(columns='v_i_p'), data_deck], axis=1)
```

```python
data['room_service'] = pd.to_numeric(data['room_service'], errors='coerce')
data['room_service'] = data['room_service'].fillna(data['room_service'].median())

data['food_court'] = pd.to_numeric(data['food_court'], errors='coerce')
data['food_court'] = data['food_court'].fillna(data['food_court'].median())

data['shopping_mall'] = pd.to_numeric(data['shopping_mall'], errors='coerce')
data['shopping_mall'] = data['shopping_mall'].fillna(data['food_court'].median())

data['spa'] = pd.to_numeric(data['spa'], errors='coerce')
data['spa'] = data['spa'].fillna(data['spa'].median())

data['v_r_deck'] = pd.to_numeric(data['v_r_deck'], errors='coerce')
data['v_r_deck'] = data['v_r_deck'].fillna(data['v_r_deck'].median())
```

```python
data['transported'] = data['transported'].astype(int)
data
```
