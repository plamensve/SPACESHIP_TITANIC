```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
data = pd.read_csv('../dataset/train.csv')
data
```

```python
data.columns = [re.sub(r'(?<!^)(?=[A-Z])', '_', col).lower() for col in data.columns]
data = data.rename(columns={
    'v_i_p': 'vip',
    'v_r_deck': 'vr_deck'
})
```

```python
data
```

```python
encoded = pd.get_dummies(data['home_planet'].fillna('Unknown')).astype(int)
```

```python
encoded
```

```python
data = pd.concat([data.drop(columns='home_planet'), encoded], axis=1)
```

```python
data
```

```python
data = data.drop(columns='passenger_id')
```

```python
data['cryo_sleep'].isna().sum()
```

```python
cryo_sleep_encoded = pd.get_dummies(data['cryo_sleep'].fillna('Unknown'), prefix='cryo_sleep').astype(int)
```

```python
data = pd.concat([data.drop(columns='cryo_sleep'), cryo_sleep_encoded], axis=1)
```

```python
data
```

```python
data['cabin']
```

```python
data[['deck', 'num', 'side']] = data['cabin'].str.split('/', expand=True)
data = data.drop(columns='cabin')
```

```python
data
```

```python
data['deck'].isna().sum()
```

```python
data_deck_encoded = pd.get_dummies(data['deck'].fillna('Unknown'), prefix='deck').astype(int)
```

```python
data_deck_encoded
```

```python
data = pd.concat([data.drop(columns='deck'), data_deck_encoded], axis=1)
```

```python
data
```

```python
data_side_encoded = pd.get_dummies(data['side'].fillna('Unknown'), prefix='side').astype(int)
data = pd.concat([data.drop(columns='side'), data_side_encoded], axis=1)
```

```python
data
```

```python
data.destination.value_counts()
```

```python
data['destination'] = data['destination'].fillna('Unknown')
```

```python
data_destination_encoded = pd.get_dummies(data['destination'], prefix='destination').astype(int)
data = pd.concat([data.drop(columns='destination'), data_destination_encoded], axis=1)
```

```python
data
```

```python
data = data.rename(columns={
    'destination_55 Cancri e': 'destination_55_cancri_e',
    'destination_PSO J318.5-22': 'destination_pso_j318_5_22',
    'destination_TRAPPIST-1e': 'destination_trappist_1_e'
})
```

```python
data
```

```python
data_vip_encoded = pd.get_dummies(data['vip'], prefix='vip').astype(int)
data = pd.concat([data.drop(columns='vip'), data_vip_encoded], axis=1)
```

```python
data
```

```python
data_transported_encoded = pd.get_dummies(data['transported'], prefix='transported').astype(int)
data = pd.concat([data.drop(columns='transported'), data_transported_encoded], axis=1)
```

```python
data.columns = data.columns.str.lower()
```

```python
data
```

```python
data.to_csv('../dataset/preprocessed_train_data.csv', index=False)
```
